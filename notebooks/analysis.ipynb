# Analysis Notebook

This notebook analyzes the results from similarity and clustering evaluations.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load results
with open('../results/similarity_results.json', 'r') as f:
    similarity_results = json.load(f)

with open('../results/clustering_results.json', 'r') as f:
    clustering_results = json.load(f)

print("Similarity Results:")
for key, value in similarity_results.items():
    print(f"{key}: {value}")

print("\nClustering Results:")
for key, value in clustering_results.items():
    print(f"{key}: {value}")
```

## Similarity Analysis

Compare Pearson and Spearman correlations for SBERT vs TF-IDF.

```python
# Plot similarity results
datasets = ['stsb', 'stsbenchmark']
methods = ['sbert', 'tfidf']
metrics = ['pearson', 'spearman']

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for i, dataset in enumerate(datasets):
    ax = axes[i]
    sbert_scores = [similarity_results[f'sbert_{dataset}'][metric] for metric in metrics]
    tfidf_scores = [similarity_results[f'tfidf_{dataset}'][metric] for metric in metrics]
    
    x = np.arange(len(metrics))
    width = 0.35
    
    ax.bar(x - width/2, sbert_scores, width, label='SBERT')
    ax.bar(x + width/2, tfidf_scores, width, label='TF-IDF')
    
    ax.set_title(f'{dataset.upper()} Similarity')
    ax.set_xticks(x)
    ax.set_xticklabels(metrics)
    ax.legend()

plt.tight_layout()
plt.show()
```
